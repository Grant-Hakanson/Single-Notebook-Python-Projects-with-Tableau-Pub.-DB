# Altering and Adding Fields

## Altering Data

```python
#Simplifying the Question Field value using a function to fill data into a new field. 

# Function:
def simpleq(x):
    if x == 'Percentage of older adults who reported subjective cognitive decline or memory loss that is happening more often or is getting worse in the preceding 12 months':
        return 'General Decline'
    elif x == 'Percentage of older adults with subjective cognitive decline or memory loss who reported talking with a health care professional about it':
        return 'Spoke to Professional'
    elif x == 'Percentage of older adults who reported that as a result of subjective cognitive decline or memory loss that they need assistance with day-to-day activities':
        return 'Impaired ADLs'
    elif x == 'Percentage of older adults who reported subjective cognitive decline or memory loss that interferes with their ability to engage in social activities or household chores':
        return 'Impaired Social and Chore Activity'
    
#New Field
cog_dcln['Affect/Stage'] = cog_dcln['Question'].apply(simpleq)
```

## Calculating Z-Scores and Removing Outliers
##### Let it be know this is a python and Data viz oriented project. If Statistics skills were meant to be shown then the approach to addressing outliers would be more thorough and precise. 

```python
# Stats info
cog_dcln.describe()
```

```python
# Convert Fields
cog_dcln['confidence_range'] = pd.to_numeric(cog_dcln['confidence_range'], errors='coerce')
cog_dcln['Data_Value'] = pd.to_numeric(cog_dcln['Data_Value'], errors='coerce')

# Fill Nulls with Median Values
cog_dcln['confidence_range'].fillna(16.6, inplace=True)
cog_dcln['Data_Value'].fillna(32.6, inplace=True)
```

```python
# create zscore fields
cog_dcln['confidence_range_z'] = stats.zscore(cog_dcln['confidence_range'])
cog_dcln['Data_Value_z'] = stats.zscore(cog_dcln['Data_Value'])

# Dropping Rows with Nulls
mask = (cog_dcln['confidence_range_z'] >= 3) | (cog_dcln['confidence_range_z'] <= -3)
rows_to_drop = cog_dcln[mask].index

mask2 = (cog_dcln['Data_Value_z'] >= 3) | (cog_dcln['Data_Value_z'] <= -3)
rows_to_drop2 = cog_dcln[mask2].index

# Drop the selected rows from the DataFrame
common_indices = rows_to_drop.intersection(rows_to_drop2)
cog_dcln.drop(common_indices, inplace=True)
rows_to_drop2 = rows_to_drop2.difference(common_indices)
cog_dcln.drop(rows_to_drop2, inplace=True)

cog_dcln.info()
```

```python
# Removing outliers from data set. 
cog_dcln['confidence_range_o'] = cog_dcln['confidence_range_z'].apply(lambda x: 'Outlier' if x >= 3 or x <= -3 else 'Normal')


cog_dcln['Data_Value_o'] = cog_dcln['Data_Value_z'].apply(lambda x: 'Outlier' if x >= 3 or x <= -3 else 'Normal')
```

## Identifying, filling, and removing nulls to optimize data for insights and visualization.

### Identifying the Nulls.

```python
# Number of Nulls for each Column

all_null_rows = pd.Series(cog_dcln.isnull().sum())

all_null_rows
```

### Filling Nulls

```python
#Making median objects.

lcl_med = cog_dcln['Low_Confidence_Limit'].median()
hcl_med = cog_dcln['High_Confidence_Limit'].median()

#Filling Nulls with median to smooth data and optimize visualization.

cog_dcln['Low_Confidence_Limit'].fillna(value=lcl_med, inplace=True)
cog_dcln['High_Confidence_Limit'].fillna(value=hcl_med, inplace=True)
```

### Removing Nulls

```python
#Removing Stratifcation Nulls
cog_dcln.dropna(subset=['Stratification Type','Stratification Value'], inplace = True)
```
