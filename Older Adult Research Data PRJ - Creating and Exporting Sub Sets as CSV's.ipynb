# Adding, Altering, and Deleting Fields
#### Why these things are done can be found in the Justification Document. [Link]

____

## Adding

```python
# Adding the Length of the study as a field. 
base['length_of_study(yrs)'] = (base['YearEnd'] - base['YearStart']) + 1
```

```python
# Adding the The Range of the confidence interval as a field. 

# convert columns to numeric type
base['Low_Confidence_Limit'] = pd.to_numeric(base['Low_Confidence_Limit'], errors='coerce')
base['High_Confidence_Limit'] = pd.to_numeric(base['High_Confidence_Limit'], errors='coerce')

# calculate range of the confidence interval
base['confidence_range'] = base['High_Confidence_Limit'] - base['Low_Confidence_Limit']
```

```python
base.head(3)
```

## Altering

```python
# Keeping the data within the Core 50 states. 

base2 = base.loc[(base['LocationDesc'] != ('Puerto Rico','Guam', 'Virgin Islands')) & 
             (base['LocationAbbr'] != ('PR','GU', 'VI'))]  
```

```python
# Adjusting field names for increased readability and visualization purposes. 
base3 = base2.rename(columns={'StratificationCategory1': 'Age Range', 
                            'StratificationCategory2': 'Stratification Type',
                            'Stratification2': 'Stratification Value',
                            'LocationDesc':'State', 'LocationAbbr':'StateAbbr', })
```

## Delete

```python
# Removing Redundant columns and empty fields. 
base4 = base3.drop(['ClassID', 'TopicID','QuestionID','LocationID','StratificationCategoryID1','StratificationID1',
                  'StratificationCategoryID2','StratificationID2','Data_Value_Alt','Sample_Size'], axis = 1)
```

# Creating Subsets and Exporting as CSVs 
### This is to for analysis of data by question connotation and one step futher by class within question type data. 

_______________________________________________

## Positive and Negative Connotation Research Questions 

```python
# Positive Connotation question split as pos_qs.
selected_questions = ['Percentage of older adults who self-reported that their health is "good", "very good", or "excellent"', 
                      'Percentage of older adults who report having lost 5 or fewer teeth due to decay or gum disease', 
                      'Percentage of older adults getting sufficient sleep (>6 hours)', 
                      'Percentage of older adults who are eating 3 or more vegetables daily', 
                      'Percentage of older adults who are eating 2 or more fruits daily']

pos_qs = base4.loc[base['Question'].isin(selected_questions)]
```

```python
# Negative Connotation question split as neg_qs.
selected_questions = ['Percentage of older adults who self-reported that their health is "good", "very good", or "excellent"', 
                      'Percentage of older adults who report having lost 5 or fewer teeth due to decay or gum disease', 
                      'Percentage of older adults getting sufficient sleep (>6 hours)', 
                      'Percentage of older adults who are eating 3 or more vegetables daily', 
                      'Percentage of older adults who are eating 2 or more fruits daily']

neg_qs = base4.loc[~base['Question'].isin(selected_questions)]
```
